Plotando a mdia mvel e o desvio padro mvel j podemos perceber que a srie no est estacionria, uma vez que o desvio padro mvel est muito distante dos dados e da mdia mvel. Vamos fazer o teste estatstico para confirmar. Importante se atentar ao p-valor, que deve ser menor que 0.05

```python
from statsmodels.tsa.stattools import adfuller
```

```python
X = df.preco_venda.values
```

```python
result = adfuller(X)

print("Teste ADF")
print(f"Teste Estatstico: {result[0]}")
print(f"P-Vaue: {result[1]}")
print(f"Valores crticos:")

for key, value in result[4].items():
  print(f"\t{key}:{value}")
```

O teste ADF (Augmented Dickey-Fuller)  uma forma comum de testar se uma srie temporal  estacionria ou no. Aqui est o que cada valor significa:

Teste Estatstico: Este  o valor que voc obteve do teste. Se o valor absoluto do teste estatstico for maior do que qualquer um dos valores crticos, ento voc pode rejeitar a hiptese nula de que a srie temporal  no-estacionria.
P-Value: Este  o valor de probabilidade associado ao seu teste estatstico. Se o p-value for menor do que o nvel de significncia (geralmente 0.05), ento voc pode rejeitar a hiptese nula.
Valores Crticos: Estes so os limites para o teste estatstico. Se o valor absoluto do seu teste estatstico for maior do que o valor crtico correspondente, ento voc pode rejeitar a hiptese nula.
No seu caso, o teste estatstico  -1.8395481433852978, e o p-value  0.36101132008604736. Como o p-value  maior do que 0.05, voc no pode rejeitar a hiptese nula, o que significa que a srie temporal  provavelmente no-estacionria. Alm disso, o valor absoluto do teste estatstico  menor do que todos os valores crticos, o que confirma essa concluso.

# Arima

Com base nos grficos de autocorrelao e autocorrelao parcial, um modelo ARIMA(1,0,0) parece ser uma boa escolha inicial. Este  um modelo autorregressivo de ordem 1 que pode ser adequado, dado o pico distinto no lag 1 nos grficos de autocorrelao e autocorrelao parcial.

```python
# Definir o tamanho do conjunto de dados de treinamento
train_size = int(len(df) * 0.8)

# Dividir os dados em treinamento e teste
train, test = df['coluna_diferenciada'][:train_size], df['coluna_diferenciada'][train_size:]
```

```python
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm

# Definir o modelo
model = ARIMA(df['coluna_diferenciada'], order=(1,0,0))

# Treinar o modelo
model_fit = model.fit()

# Imprimir o resumo do modelo
print(model_fit.summary())
```

### Vamos reverter a diferenciao para fazer previses no nvel original da srie.

Isso  importante porque as mtricas de erro que voc mencionou (MSE, MAE, RMSE) so calculadas usando os valores reais e previstos no nvel original da srie.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Fazer previses nos dados diferenciados
predictions_diff = model_fit.predict(start=train_size, end=len(df)-1)

# Reverter a diferenciao
predictions = df['preco_venda'].iloc[train_size-1] + predictions_diff.cumsum()

# Calcular o MSE, MAE e RMSE
mse = mean_squared_error(df['preco_venda'][train_size:], predictions)
mae = mean_absolute_error(df['preco_venda'][train_size:], predictions)
rmse = np.sqrt(mse)

# Calcular o Z-Score
z_scores = StandardScaler().fit_transform((df['preco_venda'][train_size:] - predictions).values.reshape(-1, 1))

# Imprimir as mtricas
print('MSE: ', mse)
print('MAE: ', mae)
print('RMSE', rmse)
print('Z-Scores: ', z_scores)
```

## Modelo de Previso STL (Seasonal and Trend decomposition using Loess)

O modelo STL  um procedimento para decompor uma srie temporal em trs componentes: tendncia, sazonalidade e resduo. Ele  frequentemente combinado com outros modelos de previso, como ARIMA, para lidar com a parte residual da srie temporal.

```python
from statsmodels.tsa.forecasting.stl import STLForecast
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np

# Definir o tamanho do conjunto de dados de treinamento
train_size = int(len(df) * 0.8)

# Dividir os dados em treinamento e teste
train, test = df['preco_venda'][:train_size], df['preco_venda'][train_size:]

# Definir o modelo
model_STL = STLForecast(train, ARIMA, model_kwargs={"order": (1, 0, 0)})

# Treinar o modelo
model_STL_fit = model_STL.fit()

# Fazer previso
forecast_STL = model_STL_fit.forecast(steps=len(test))

# Calcular o WMAPE
wmape = np.sum(np.abs(forecast_STL - test)) / np.sum(test)
print(f"WMAPE: {wmape:.2%}")

# Criar a figura
fig = go.Figure()

# Adicionar os dados de treinamento e teste ao grfico
fig.add_trace(go.Scatter(x=train.index, y=train, mode='lines', name='Treinamento'))
fig.add_trace(go.Scatter(x=test.index, y=test, mode='lines', name='Teste'))

# Adicionar as previses ao grfico
fig.add_trace(go.Scatter(x=test.index, y=forecast_STL, mode='lines', name='Previso'))

# Atualizar o layout
fig.update_layout(title='Previso usando STL Forecasting',
                  xaxis_title='Data',
                  yaxis_title='Valor')

# Mostrar o grfico
fig.show()
```

O STL forecasting tambm deu um um wmape dentro do padro bom. Dentro do plot acima, podemos perceber que a previso est um pouco mais prxima da realidade. Mas vamos buscar um resultado melhor.
