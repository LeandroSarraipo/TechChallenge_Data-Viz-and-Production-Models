Vamos então verificar se a série está estacionária, algo que é essencial a alguns modelos.

```python
rolmean = df.rolling(window=12).mean() #média móvel em 12 meses (dando a média móvel no nível anual).
rolstd = df.rolling(window=12).std()   #Desvio padrão em 12 meses
print(rolmean,rolstd)
```

```python
orig = plt.plot(df, color='blue', label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
std = plt.plot(rolstd, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Média móvel e desvio padrão')
plt.show(block=False)
```

Plotando a média móvel e o desvio padrão móvel já podemos perceber que a série não está estacionária, uma vez que o desvio padrão móvel está muito distante dos dados e da média móvel. Vamos fazer o teste estatístico para confirmar. Importante se atentar ao p-valor, que deve ser menor que 0.05

```python
from statsmodels.tsa.stattools import adfuller
```

```python
X = df.preco_venda.values
```

```python
result = adfuller(X)

print("Teste ADF")
print(f"Teste Estatístico: {result[0]}")
print(f"P-Vaue: {result[1]}")
print(f"Valores críticos:")

for key, value in result[4].items():
  print(f"\t{key}:{value}")
```

O teste ADF (Augmented Dickey-Fuller) é uma forma comum de testar se uma série temporal é estacionária ou não. Aqui está o que cada valor significa:

Teste Estatístico: Este é o valor que você obteve do teste. Se o valor absoluto do teste estatístico for maior do que qualquer um dos valores críticos, então você pode rejeitar a hipótese nula de que a série temporal é não-estacionária.
P-Value: Este é o valor de probabilidade associado ao seu teste estatístico. Se o p-value for menor do que o nível de significância (geralmente 0.05), então você pode rejeitar a hipótese nula.
Valores Críticos: Estes são os limites para o teste estatístico. Se o valor absoluto do seu teste estatístico for maior do que o valor crítico correspondente, então você pode rejeitar a hipótese nula.
No seu caso, o teste estatístico é -1.8395481433852978, e o p-value é 0.36101132008604736. Como o p-value é maior do que 0.05, você não pode rejeitar a hipótese nula, o que significa que a série temporal é provavelmente não-estacionária. Além disso, o valor absoluto do teste estatístico é menor do que todos os valores críticos, o que confirma essa conclusão.

Valores Críticos: Como mencionado, o valor absoluto do seu teste estatístico é maior do que todos os valores críticos, o que é outro indicador de que a série é estacionária.

# Arima

Com base nos gráficos de autocorrelação e autocorrelação parcial, um modelo ARIMA(1,0,0) parece ser uma boa escolha inicial. Este é um modelo autorregressivo de ordem 1 que pode ser adequado, dado o pico distinto no lag 1 nos gráficos de autocorrelação e autocorrelação parcial.

```python
# Definir o tamanho do conjunto de dados de treinamento
train_size = int(len(df) * 0.8)

# Dividir os dados em treinamento e teste
train, test = df['coluna_diferenciada'][:train_size], df['coluna_diferenciada'][train_size:]
```

```python
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm

# Definir o modelo
model = ARIMA(df['coluna_diferenciada'], order=(1,0,0))

# Treinar o modelo
model_fit = model.fit()

# Imprimir o resumo do modelo
print(model_fit.summary())
```

### Vamos reverter a diferenciação para fazer previsões no nível original da série.

Isso é importante porque as métricas de erro que você mencionou (MSE, MAE, RMSE) são calculadas usando os valores reais e previstos no nível original da série.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Fazer previsões nos dados diferenciados
predictions_diff = model_fit.predict(start=train_size, end=len(df)-1)

# Reverter a diferenciação
predictions = df['preco_venda'].iloc[train_size-1] + predictions_diff.cumsum()

# Calcular o MSE, MAE e RMSE
mse = mean_squared_error(df['preco_venda'][train_size:], predictions)
mae = mean_absolute_error(df['preco_venda'][train_size:], predictions)
rmse = np.sqrt(mse)

# Calcular o Z-Score
z_scores = StandardScaler().fit_transform((df['preco_venda'][train_size:] - predictions).values.reshape(-1, 1))

# Imprimir as métricas
print('MSE: ', mse)
print('MAE: ', mae)
print('RMSE', rmse)
print('Z-Scores: ', z_scores)
```

## Modelo de Previsão STL (Seasonal and Trend decomposition using Loess)

O modelo STL é um procedimento para decompor uma série temporal em três componentes: tendência, sazonalidade e resíduo. Ele é frequentemente combinado com outros modelos de previsão, como ARIMA, para lidar com a parte residual da série temporal.

```python
from statsmodels.tsa.forecasting.stl import STLForecast
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np

# Definir o tamanho do conjunto de dados de treinamento
train_size = int(len(df) * 0.8)

# Dividir os dados em treinamento e teste
train, test = df['preco_venda'][:train_size], df['preco_venda'][train_size:]

# Definir o modelo
model_STL = STLForecast(train, ARIMA, model_kwargs={"order": (1, 0, 0)})

# Treinar o modelo
model_STL_fit = model_STL.fit()

# Fazer previsão
forecast_STL = model_STL_fit.forecast(steps=len(test))

# Calcular o WMAPE
wmape = np.sum(np.abs(forecast_STL - test)) / np.sum(test)
print(f"WMAPE: {wmape:.2%}")

# Criar a figura
fig = go.Figure()

# Adicionar os dados de treinamento e teste ao gráfico
fig.add_trace(go.Scatter(x=train.index, y=train, mode='lines', name='Treinamento'))
fig.add_trace(go.Scatter(x=test.index, y=test, mode='lines', name='Teste'))

# Adicionar as previsões ao gráfico
fig.add_trace(go.Scatter(x=test.index, y=forecast_STL, mode='lines', name='Previsão'))

# Atualizar o layout
fig.update_layout(title='Previsão usando STL Forecasting',
                  xaxis_title='Data',
                  yaxis_title='Valor')

# Mostrar o gráfico
fig.show()
```

O STL forecasting também deu um um wmape dentro do padrão bom. Dentro do plot acima, podemos perceber que a previsão está um pouco mais próxima da realidade. Mas vamos buscar um resultado melhor.

```python
# Extrair as colunas relevantes dos DataFrames
previsao_cols_2 = ['ds', 'yhat']
valores_reais_cols_2 = ['ds', 'y']

previsao_2 = previsao_2[previsao_cols_2]
valores_reais_2 = train_data_2[valores_reais_cols_2]

# Mesclar os DataFrames nas colunas 'ds' para comparar previsões e valores reais
resultados_2 = pd.merge(previsao_2, valores_reais_2, on='ds', how='inner')

# Calcular o erro percentual absoluto para cada ponto de dados
resultados_2['erro_percentual_absoluto'] = np.abs((resultados_2['y'] - resultados_2['yhat']) / resultados_2['y']) * 100

# Calcular o MAPE
mape_2 = np.mean(resultados_2['erro_percentual_absoluto'])

print(f"MAPE: {mape_2:.2f}%")
```
